# Seaborn.heatmap - properties
seaborn.heatmap(), you typically need to pass in a matrix of data. 
Then, you can adjust the parameters to customize your heatmaps depending on your requirements.

data: The dataset to visualize, which must be in a matrix form.
cmap: Specifies the colormap for the heatmap. Seaborn supports various color palettes, 
    including sequential, diverging, and qualitative schemes.
annot: If set to True, the value in each cell is annotated on the heatmap.
fmt: When annot is True, fmt determines the string formatting code for annotating the data.
    For example, 'd' for integers and '.2f' for floating-point numbers with two decimals.
linewidths: Sets the width of the lines that will divide each cell. A larger value 
    increases the separation between cells.
linecolor: Specifies the color of the lines that divide each cell if linewidths is greater than 0.
cbar: A boolean value that indicates whether to draw a color bar. The color bar 
    provides a reference for mapping data values to colors.
vmin and vmax: These parameters define the data range that the colormap covers.
center: Sets the value at which to center the colormap when using diverging color schemes.
square: If set to True, ensures that the heatmap cells are square-shaped.
xticklabels and yticklabels: Control the labels displayed on the x and y axes.

```python
# Generate data randomly
data = np.random.randint(low = 1, 
                         high = 100, 
                         size = (10, 10))
```

```python
data
```

```python
# Define column names
column_names = [f"Col_{i}" for i in range(1, 11)]
```

```python
column_names
```

```python
# Convert the NumPy array to a DataFrame
df = pd.DataFrame(data, columns=column_names)
df
```

```python
correlation_matrix = df.corr()
```

```python
correlation_matrix
```

```python
# Create the heatmap
plt.figure(figsize = (10,8))
sns.heatmap(correlation_matrix, cmap = 'coolwarm')
plt.show()
```

```python
# Create the heatmap
plt.figure(figsize = (10,8))
sns.heatmap(correlation_matrix, cmap = 'PRGn')
plt.show()
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(correlation_matrix,
            cmap = 'PRGn',
            vmin = -1, vmax = 1, center = 0)
plt.show()
```

```python
# Create an annotated heatmap
plt.figure(figsize = (10,8))
plt.rcParams.update({'font.size': 12})
sns.heatmap(correlation_matrix,
            cmap = 'PRGn',
            vmin = -1, vmax = 1, center = 0,
            annot=True, fmt=".2f",
            square=True,
            linewidths=.5)
plt.show()
```

```python
# Create an annotated heatmap
plt.figure(figsize = (10,8))
plt.rcParams.update({'font.size': 12})
sns.heatmap(correlation_matrix,
            cmap = 'PRGn',
            vmin = -1, vmax = 1, center = 0,
            annot=True, fmt=".2f",
            square=True,
            linewidths=.5,
            linecolor='red')
plt.show()
```

```python
# Create a mask using numpy's triu function
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Create an annotated heatmap
plt.figure(figsize = (10,8))
plt.rcParams.update({'font.size': 12})
sns.heatmap(correlation_matrix,
            cmap = 'PRGn',
            vmin = -1, vmax = 1, center = 0,
            annot=True, fmt=".2f",
            square=True,
            linewidths=.5,
           mask=mask)
plt.show()
```

```python
# Create an annotated heatmap
plt.figure(figsize = (10,8))
plt.rcParams.update({'font.size': 12})
ax = sns.heatmap(correlation_matrix,
            cmap = 'PRGn',
            vmin = -1, vmax = 1, center = 0,
            annot=True, fmt=".1f",
            square=True,
            linewidths=.5)
ax.set(xlabel="", ylabel="")
ax.xaxis.tick_top()
plt.show()
```

#  Cluster - Heatmap

```python
iris = sns.load_dataset("iris")
iris
```

```python
species = iris.pop("species")
iris
```

```python
sns.clustermap(iris)
```

```python
sns.clustermap(
    iris,
    figsize=(7, 5),
    row_cluster=False,
    col_cluster=True,
    dendrogram_ratio=(.1, .2),
    cbar_pos=(0, .2, .03, .4)
)
```

```python
sns.clustermap(
    iris,
    figsize=(7, 5),
    row_cluster=True,
    col_cluster=False,
    dendrogram_ratio=(.1, .2),
    cbar_pos=(1.0, .2, .03, .4)
)
```

```python
lab = dict(zip(species.unique(), "rbg"))
lab
```

```python
row_colors = species.map(lab)
row_colors
```

```python
sns.clustermap(iris, row_colors=row_colors)
```

```python
sns.clustermap(iris,
               figsize=(7,7),
               row_colors=row_colors,
               col_cluster=False,)
```

```python
# Create the clustermap
g = sns.clustermap(iris,
                   figsize=(7, 7),
                   row_colors=row_colors,
                   col_cluster=False)

# Add a legend manually
for label in lab:
    g.ax_row_dendrogram.bar(5, 5,
                            color=lab[label],
                            label=label,
                            linewidth=0)
g.ax_row_dendrogram.legend(title='Species',loc='center', bbox_to_anchor=(3, 1.1), ncol=3)

plt.show()
```

# Z-Score Based Clustering
The Z-score is a statistical measure that describes a value's relationship to the mean of a group of values. 
It is measured in terms of standard deviations from the mean.

z_score Parameter in Seaborn's clustermap
The z_score parameter in sns.clustermap allows you to normalize the data along a specific axis before clustering:

z_score=0: Normalizes along the rows (default). Each feature (column) in a row will have mean 0 and standard deviation 1.

z_score=1: Normalizes along the columns. Each observation (row) in a column will have mean 0 and standard deviation 1.

```python
# Z-Score based clustering
sns.clustermap(iris,
               figsize=(5, 5),
               z_score=0,
               cmap="vlag",
               center=0)
```

```python
sns.clustermap(iris,
               figsize=(5, 5),
               z_score=1,
               cmap="vlag",
               center=0)
```
