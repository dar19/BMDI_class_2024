```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings

# Suppress specific warnings
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
#create DataFrame
df = pd.DataFrame({'date': ['1/1/2021',
                            '1/2/2021',
                            '1/3/2021',
                            '1/4/2021',
                            '1/1/2021',
                            '1/2/2021',
                            '1/3/2021',
                            '1/4/2021'],
                   'sales': [4, 7, 8, 13, 17, 15, 21, 28],
                   'company': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B']})
```

```python
#plot multiple time series
sns.lineplot(x='date', y='sales', hue='company', data=df)
```

```python
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")
```

```python
fmri
```

```python
# Plot the responses for different regions
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             data=fmri)
```

```python
# Plot the responses for different events
sns.lineplot(x="timepoint",
             y="signal",
             hue="event",
             data=fmri)
```

```python
# Plot the responses for different events and regions
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             style="event",
             data=fmri)
```
